```python
bins=[0,100000,300000,float('inf')]
labels=['low','med','high']
df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
df['Income Band'].value_counts().plot(kind='bar')
```

```python
categorical_vals=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]

for col in categorical_vals:
    print(f"Value counts for '{col}':")
    display(df[col].value_counts())
```

we analyze every column using univariate and bivariate analysis for drawing the insights from it which is the most important and needed step in data analytics

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance']

plt.figure(figsize=(10,6))
for i,col in enumerate(numerical_cols):
    plt.subplot(3,1,i+1)
    sns.histplot(df[col],kde=True)
plt.tight_layout()
plt.show()
```

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts']

correlation_matrix=df[numerical_cols].corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```
