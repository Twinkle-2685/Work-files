```python
weight=int(input("Enter your weight in pounds: "))

height=int(input("Enter your height in inches:"))

BMI = (weight*703)/(height*height)

print(round(BMI,2))

if BMI>0:
    if BMI<18.5:
        print("Underweight")
    elif BMI<=24.9:
        print("Normal weight")
    elif BMI<=29.9:
        print("Normal weight")
    elif BMI<=34.9:
        print("Normal weight")
    elif BMI<=39.9:
        print("Normal weight")
    else:
        print("Morbedely obese")
else:
    print("Enter valid inputs")
```
